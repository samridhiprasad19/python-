```python
class myiter:
    def __init__(self,num,limit):
        self.num = num
        self.limit = limit
    def __iter__(self):
        self.x = 1
        return self
    def __next__(self):
        if self.x>self.limit:
            self.x = self.x +1
            raise StopIteration #to show error
        else:
            self.num = self.num+1
            self.x = self.x+1
            return self.num
        
        
```

```python
k = myiter(10,100)
k = iter(k)
```

```python
while True:
    print(next(k),end = "\t")
```

```python
#iterator to print infinite even number
class even_odd:
    def __iter__(self):
        self.x = 1
        return self
    
    def __next__(self):
        while True:
            if self.x % 2 == 0:
                print("Even = {}".format(self.x))
            else:
                print("Odd = {}".format(self.x))
            self.x = self.x + 1
        
        
        
        
```

```python
k = even_odd()
k = iter(k)
k = next(k)
```
